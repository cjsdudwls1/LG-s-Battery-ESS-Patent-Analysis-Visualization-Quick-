## 2. 단계별 분석 방법 및 코드 로직

### ① 데이터 전처리 (Preprocessing)
원천 데이터의 노이즈를 제거하고 분석 가능한 형태로 가공하는 핵심 단계.

* **날짜 데이터 변환:**
    * **코드 로직:** `pd.to_datetime(..., errors='coerce')`
    * **목적:** 문자열(`2023.05.01`)을 시계열 분석이 가능한 `datetime` 객체로 변환. 오류 데이터는 `NaT`로 처리.
* **IPC(국제특허분류) 코드 단순화:**
    * **코드 로직:** `text.split('|')[0].split('/')[0]`
    * **목적:** `H01M 10/052(2010.01)` 처럼 복잡한 코드를 메인 서브클래스(`H01M 10`) 단위로 잘라서 기술 분야를 그룹화함.
* **출원인/발명자 텍스트 정제:**
    * **코드 로직:** `re.sub(r'\([^)]*\)', '', text)`
    * **목적:** 이름 뒤의 주소정보(`(서울특별시...)`)를 정규표현식으로 제거하여 동일 인물/기업을 정확하게 카운팅함.

### ② 시계열 트렌드 분석 (Time Series Analysis)
기술의 성장 단계(태동기-성장기-성숙기)를 파악하기 위한 분석.

* **분석 방법:** 출원일자와 등록일자 기준 연도별 건수 집계 (`value_counts().sort_index()`).
* **시각화:** 꺾은선 그래프(Line Plot)를 통해 출원량의 우상향 추세 확인.
* **인사이트:** 최근 3~4년 내 급격한 출원 증가는 기술 투자가 집중되고 있음을 의미.

### ③ 기술 분야 및 법적 상태 분석 (Technology & Legal Status)
어떤 기술에 집중하고 있으며, 권리 확보율은 얼마나 되는지 분석.

* **분석 방법:** 전처리된 `IPC_Main` 컬럼의 빈도수 상위 10개 추출.
* **시각화:** `sns.countplot`을 사용하여 기술 분야(X축)별 출원 건수(Y축)를 표시하되, `hue='법적상태'` 옵션으로 등록/공개 비율을 동시에 시각화.
* **핵심 발견:** `H01M 50` (구조/팩/케이스) 분야가 `H01M 10` (소재/셀) 분야보다 압도적으로 많음 → **패키징 기술 중심의 시장 재편.**

### ④ 핵심 발명자-기술 매핑 (Inventor-Tech Heatmap)
"누가(Who) 무엇을(What) 연구하는가?"를 규명하는 고난이도 분석.

* **데이터 구조화:** 1.  한 특허에 여러 발명자가 있는 경우 `split('|')`로 분리하여 행(Row)을 늘림 (Explode).
    2.  발명자별 `IPC_Code` 빈도를 카운팅.
* **히트맵(Heatmap) 구현:**
    * `pd.crosstab`으로 발명자(행) × 기술분야(열) 교차표 생성.
    * **IPC 라벨링:** 암호 같은 코드를 `'H01M 50\n(구조/케이스/팩)'` 형태의 직관적인 한글로 변환하는 매핑 딕셔너리(`ipc_meaning_map`) 적용.
* **인사이트:** 상위 10명의 에이스 연구원들이 모두 화학 분야가 아닌 **기구 설계(H01M 50)에 집중**하고 있음을 시각적으로 입증.

### ⑤ 텍스트 마이닝 (WordCloud)
특허 제목에서 빈출 키워드를 추출하여 직관적인 이미지를 생성.

* **불용어(Stopwords) 처리 강화:**
    * 일반적인 불용어 외에 특허 상투어(`조성물`, `방법`, `장치`, `제공`) 및 데이터셋 공통어(`배터리`, `이차전지`)를 제거 리스트에 추가.

## 4. [심층 분석] 1등 특허 (가장 많이 인용된 기술)
**"LG 배터리 모듈 기술의 정수: 마찰 교반 용접(FSW)을 적용한 케이스"**

### ① 서지 정보
* **발명의 명칭:** 배터리 모듈 (Battery Module)
* **출원 번호:** 10-2016-0122320
* **피인용 횟수:** 22회 (해당 데이터셋 내 1위, 기술적 파급력이 매우 큼)
* **기술 분류:** H01M 10/6554 (냉각 구조), H01M 50 (구조)

### ② 핵심 기술 요약
* **기술 개요:** 다수의 파우치형 전지를 감싸는 모듈 케이스를 조립할 때, 나사나 접착제 대신 **'마찰 교반 용접(FSW)'** 기술을 사용하여 부품을 일체화하는 기술.
* **작동 원리 (쉬운 설명):**
    * 금속을 녹이지 않고 고속으로 비벼서 발생하는 마찰열로 접합함 (마치 찰흙 두 덩이를 문질러 하나로 만드는 것과 유사).
    * 기존 용접 대비 변형이 적고, 접합 강도가 매우 높으며 밀폐력이 뛰어남.

### ③ 기술적 특징 (Key Features)
* **6면체 모듈 케이스 구조:**
    *
    * 쿨링 플레이트(바닥) + 셀 어셈블리(배터리 뭉치) + 커버(전/후/좌/우/상)의 완벽한 결합.
* **마찰 교반 용접(FSW) 적용:**
    * 모든 케이스 부품 간의 연결에 FSW를 적용하여 체결 부품(볼트/너트)을 제거, 경량화 및 공간 효율성 극대화.
* **맞대기 용접(Butt Joint) 설계:**
    * 부재를 겹치지 않고 모서리를 맞대어 용접함으로써 전체 모듈의 부피를 줄이고 매끄러운 외관 확보.

### ④ 권리 범위 및 가치 평가
* **특허 성격:** 특정 제조 공법(FSW)과 조립 구조를 한정한 **'공정 및 구조 개선 특허'**.
* **권리 범위:**
    * 광범위한 원천 특허라기보다는, **'FSW 용접을 사용한 특정 형태의 파우치형 모듈'**로 한정됨.
    * 하지만 경쟁사가 동일한 방식(FSW)으로 고품질 모듈을 제조하려 할 때 반드시 회피해야 하는 강력한 길목 특허임.
* **시장 가치:** 피인용 수가 높은 이유는 이 기술이 배터리 팩의 **경량화**, **고강도화**, **방수/방진(밀폐)** 문제를 동시에 해결하는 가장 효율적인 솔루션 중 하나이기 때문임.

```python
# ==========================================
# [0] 라이브러리 및 폰트 설정 (Colab용)
# ==========================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
import re
import google.generativeai as genai
from google.colab import userdata

# 한글 폰트 설정
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
sns.set(font='NanumBarunGothic', style='whitegrid')

# ==========================================
# [1] 데이터 전처리
# ==========================================
def preprocess_patent_data(df):
    df = df.copy()

    # 1. 날짜 컬럼 변환
    date_cols = ['출원일자', '등록일자', '공개일자', '공고일자']
    for col in date_cols:
        df[col] = pd.to_datetime(df[col], errors='coerce')

    # 2. IPC 메인 그룹 추출 (Subclass 레벨)
    def extract_main_code(text):
        if pd.isna(text): return "Unknown"
        first_code = text.split('|')[0]
        return first_code.split('/')[0]

    df['IPC_Main'] = df['IPC분류'].apply(extract_main_code)
    return df

# ★ 데이터 로드 (df_filtered가 이미 있다고 가정)
# 만약 엑셀파일을 다시 읽어야 한다면 아래 주석 해제
# df_filtered = pd.read_excel('/content/경로/파일명.xlsx')

df_clean = preprocess_patent_data(df_filtered)

# ==========================================
# [2] 시계열 분석 (출원/등록 추이)
# ==========================================
df_clean['출원년도'] = df_clean['출원일자'].dt.year
df_clean['등록년도'] = df_clean['등록일자'].dt.year

app_trend = df_clean['출원년도'].value_counts().sort_index()
reg_trend = df_clean['등록년도'].value_counts().sort_index()

plt.figure(figsize=(14, 5))
plt.plot(app_trend.index, app_trend.values, marker='o', label='출원 건수', linewidth=2)
plt.plot(reg_trend.index, reg_trend.values, marker='s', linestyle='--', label='등록 건수', linewidth=2)
plt.title('연도별 특허 출원 및 등록 추이', fontsize=16)
plt.xlabel('연도')
plt.ylabel('건수')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# ==========================================
# [3] IPC 기술 분야 분석
# ==========================================
print("\n### 핵심 기술 분야(IPC Main) 상위 10개 ###")
top_ipc = df_clean['IPC_Main'].value_counts().head(10).index
df_ipc_top = df_clean[df_clean['IPC_Main'].isin(top_ipc)]

plt.figure(figsize=(14, 7))
sns.countplot(data=df_ipc_top, x='IPC_Main', hue='법적상태', order=top_ipc, palette='viridis')
plt.title('주요 기술 분야별 특허 법적 상태', fontsize=18)
plt.xlabel('IPC 기술 분류', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# ==========================================
# [4] 출원인 분석 (괄호 제거)
# ==========================================
all_applicants = []
for item in df_clean['출원인']:
    if pd.notnull(item):
        cleaned = re.sub(r'\([^)]*\)', '', item) # 괄호 제거
        all_applicants.append(cleaned.strip())

top_applicants = pd.Series(all_applicants).value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_applicants.values, y=top_applicants.index, palette='magma')
plt.title('상위 10개 출원인', fontsize=15)
plt.show()

# ==========================================
# [5] 핵심 발명자 히트맵 (★ 발명자별 특허 총 개수 열 추가됨)
# ==========================================
# 1. IPC 코드 해석 사전
ipc_meaning_map = {
    'H01M 50': 'H01M 50\n(구조/케이스/팩)',
    'H01M 10': 'H01M 10\n(이차전지 일반)',
    'H01M 4':  'H01M 4\n(전극/활물질)',
    'H01M 52': 'H01M 52\n(스택/적층구조)',
    'B60L 50': 'B60L 50\n(전기차 추진)',
    'B60L 53': 'B60L 53\n(충전 시스템)',
    'B60L 58': 'B60L 58\n(BMS/제어)',
    'G01R 31': 'G01R 31\n(테스트/검사)',
    'H01G 11': 'H01G 11\n(슈퍼커패시터)',
    'H01M 8':  'H01M 8\n(연료전지)'
}

# 2. 데이터 준비
inventor_tech = []
for idx, row in df_clean.iterrows():
    if pd.isna(row['발명자']): continue

    raw_inventors = re.sub(r'\([^)]*\)', '', row['발명자'])
    inventors = [inv.strip() for inv in raw_inventors.split('|')]
    ipc_code = row['IPC_Main'].strip()

    for inv in inventors:
        inventor_tech.append({'발명자': inv, 'IPC_Code': ipc_code})

df_inv_tech = pd.DataFrame(inventor_tech)

# ⭐️ 추가 코드 1: 발명자별 특허 총 개수 계산
patent_counts = df_inv_tech['발명자'].value_counts().rename('특허 총 개수')
df_patent_counts = patent_counts.reset_index().rename(columns={'index': '발명자'})

# 3. 상위 추출 및 필터링
top_inventors = df_inv_tech['발명자'].value_counts().head(10).index
top_ipcs = df_inv_tech['IPC_Code'].value_counts().head(10).index

# 상위 10개 발명자 목록에 해당하는 특허 개수 데이터프레임 필터링
df_top_counts = df_patent_counts[df_patent_counts['발명자'].isin(top_inventors)]


df_filtered_heatmap = df_inv_tech[
    (df_inv_tech['발명자'].isin(top_inventors)) &
    (df_inv_tech['IPC_Code'].isin(top_ipcs))
]

# 4. IPC 라벨 변환 및 시각화 (수정)
df_filtered_heatmap['IPC_Label'] = df_filtered_heatmap['IPC_Code'].apply(
    lambda x: ipc_meaning_map.get(x, x)
)
heatmap_data = pd.crosstab(df_filtered_heatmap['발명자'], df_filtered_heatmap['IPC_Label'])

# ⭐️ 추가 코드 2: 히트맵 데이터와 특허 총 개수 데이터 병합
# 발명자를 인덱스로 설정
heatmap_data = heatmap_data.reset_index().set_index('발명자')
df_top_counts = df_top_counts.set_index('발명자')

# 두 데이터프레임을 병합 (특허 총 개수 열 추가)
merged_heatmap_data = pd.merge(
    df_top_counts[['특허 총 개수']], # 총 개수 열만 선택
    heatmap_data,
    left_index=True,
    right_index=True,
    how='inner'
)
merged_heatmap_data = merged_heatmap_data.sort_values(by='특허 총 개수', ascending=False) # 총 개수 기준으로 정렬

# 시각화
plt.figure(figsize=(16, 9)) # figsize를 약간 늘려줍니다.
sns.heatmap(merged_heatmap_data, annot=True, fmt='d', cmap='YlGnBu', linewidths=.5)
plt.title('핵심 발명자(Key Man)의 주력 기술 분야 (총 개수 포함)', fontsize=18)
plt.xlabel('기술 분야 (IPC)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.ylabel('발명자', fontsize=14)
plt.tight_layout()
plt.show()

# ==========================================
# [6] 1등 특허 식별 (피인용 횟수 기준)
# ==========================================
top_cited_patent = df_clean.sort_values(by='피인용 횟수', ascending=False).head(1)
best_claim = ""
best_title = ""

print("\n### 🏆 가장 많이 인용된 1등 특허 정보 ###")
if not top_cited_patent.empty:
    # display()는 Colab 환경에서 동작하는 함수이며, 일반 파이썬 환경에서는 print로 대체해야 합니다.
    # 여기서는 제공된 코드 구조를 유지합니다.
    try:
        from IPython.display import display
        display(top_cited_patent[['발명의명칭', '출원번호', '피인용 횟수', 'IPC분류']].T)
    except ImportError:
        print(top_cited_patent[['발명의명칭', '출원번호', '피인용 횟수', 'IPC분류']].T)

    best_claim = top_cited_patent['청구항'].values[0]
    best_title = top_cited_patent['발명의명칭'].values[0]
else:
    print("피인용 횟수 데이터가 없습니다.")

# ==========================================
# [7] 워드클라우드 (★ 강력 필터링 적용됨)
# ==========================================
text_corpus = ' '.join(df_clean['발명의명칭'].astype(str))

# 1. 불필요한 문구 강제 삭제
text_corpus = text_corpus.replace('이를 포함하는', '')
text_corpus = text_corpus.replace('이를', '')
text_corpus = text_corpus.replace('포함하는', '')
text_corpus = re.sub(r'\b그\b', '', text_corpus) # '그' 단독 제거
text_corpus = re.sub(r'\S+하는', '', text_corpus) # '~하는'으로 끝나는 어절 제거

# 2. 불용어 리스트
stopwords = {
    '발명의', '명칭', '방법', '장치', '시스템', '제조', '단계', '특징', '구비', '제공', '조성물', '사용', '이용',
    '및', '위한', '또는', '통한', '관한', '포함', '상기', '의', '에', '를', '은', '는', '이', '가',
    '배터리', '전지', '이차전지', '모듈', '팩', '셀'
}

font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'

try:
    wordcloud = WordCloud(
        font_path=font_path,
        width=1000, height=500,
        background_color='white',
        colormap='Dark2',
        stopwords=stopwords,
        regexp=r"[가-힣a-zA-Z]+"
    ).generate(text_corpus)

    plt.figure(figsize=(12, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('핵심 기술 키워드 (불용어 제거됨)', fontsize=16)
    plt.show()
except Exception as e:
    print(f"워드클라우드 오류: {e}")

# ==========================================
# [8] Gemini API 분석 (최신 모델)
# ==========================================
print("\n========== [Gemini AI 특허 분석 시작] ==========")

if best_claim:
    try:
        # Colab 환경에서 실행하려면 'Gemini_API_KEY'를 Secrets에 저장해야 합니다.
        GOOGLE_API_KEY = userdata.get('Gemini_API_KEY')
        genai.configure(api_key=GOOGLE_API_KEY)

        # 모델: gemini-2.5-flash
        model = genai.GenerativeModel('gemini-2.5-flash')

        patent_claim_short = best_claim[:3000]

        prompt = f'''
        당신은 20년 경력의 배터리 분야 전문 변리사입니다.
        아래 특허의 [청구항]을 분석하여 다음 내용을 작성해주세요.

        1. [핵심 요약]: 이 기술이 무엇인지 한 문장으로 요약하세요.
        2. [쉬운 설명]: 기존 기술과 비교했을 때 어떤 점이 혁신적인지 초등학생도 이해할 수 있게 설명하세요.
        3. [기술적 특징]: 이 특허가 보호하고자 하는 핵심 구성요소 3가지를 글머리 기호로 나열하세요.
        4. [권리 범위 평가]: 청구항의 구조를 보았을 때 권리 범위가 넓은 원천 기술인지, 특정 구현에 한정된 기술인지 평가하세요.

        [특허 정보]
        - 명칭: {best_title}
        - 청구항: {patent_claim_short}...
        '''

        print(f"분석 대상: {best_title}")
        print("Gemini가 분석 중입니다... (모델: gemini-2.5-flash)\n")
        response = model.generate_content(prompt)
        print("--------- [Gemini 분석 결과] ---------")
        print(response.text)
        print("--------------------------------------")

    except Exception as e:
        print(f"❌ Gemini API 오류: {e}")
        print("👉 Google Colab '비밀(Secrets)'에 'Gemini_API_KEY'가 있는지 확인하세요.")
else:
    print("분석할 1등 특허가 없습니다.")
```
